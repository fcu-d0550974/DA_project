```python
import csv
import json
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./CSV2JSON.csv')
df = pd.DataFrame(df)
member = df.groupby(['member_id'])

id = df['member_id'].unique()
```

```python
# get tag_name

file = open('./CSV2JSON.csv', 'r', encoding = 'utf-8-sig')
f = csv.DictReader(file)

l = []
for i in f:
    d = {}
    for k,v in i.items():
        d[k] = v
    l.append(d)
    
    
tag = []
for n in range(len(id)):
    for j in range(len(l)):
            
        if id[n] == l[j]['member_id']:
            tag.append({'id': id[n], 'tag_name': l[j]['tag_name']})
        
        if len(tag) > 2:
            if tag[-1] == tag[-2] or tag[-1] == tag[-3]:
                tag.pop(-1)
            elif tag[-2] == tag[-3]:
                tag.pop(-2)
```

```python
print(l)
```

```python
data = []

for i in range(len(id)):
    data.append({'_id': id[i], 'member_id': id[i]})

for j in range(len(data)):
    tags = []
    for k in range(len(tag)):
        if data[j]['_id'] == tag[k]['id']:
            tags.append({'tag_name': tag[k]['tag_name']})

    data[j]['tags'] = tags
```

```python
for i in range(len(data)):
    
    for x in range(len(data[i]['tags'])):
        detail = []
        for h in range(len(l)):
            if data[i]['_id'] == l[h]['member_id'] and data[i]['tags'][x]['tag_name'] == l[h]['tag_name']:
                detail.append({'detail_name': l[h]['detail_name'], 'detail_value': l[h]['detail_value']})
            

        data[i]['tags'][x]['detail'] = detail
```
